```python
import numpy as np
import matplotlib.pyplot as plt
```

En este cuaderno veremos algunos ejemplos de los tipos de Series de tiempo que se pueden observar con sus diferentes componenetes.

En primer Lugar se ve cómo podemos ver una Serie de tiempo con Tendencia Lineal

```python
# Generar datos de serie de tiempo con tendencia lineal
np.random.seed(0)
n = 100  # Número de puntos de datos
t = np.arange(n)  # Tiempo
slope = 0.5  # Pendiente de la tendencia lineal
intercept = 10  # Intersección de la tendencia lineal
epsilon = np.random.normal(scale=5, size=n)  # Error aleatorio
```

Despúes de generar los datos pertinentes para la serie de tiempo, entonces podemos visualizar su fórmula, que está dada por los parámetros establecidos en el código anterior

```python
# Serie de tiempo con tendencia lineal: Y_t = intercept + slope * t + epsilon
y = intercept + slope * t + epsilon
```

Se procede a graficar la respectiva serie.

```python
# Graficar la serie de tiempo
plt.figure(figsize=(10, 6))
plt.plot(t, y, label='Serie de tiempo lineal', color='blue')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.title('Serie de tiempo con tendencia lineal')
plt.legend()
plt.grid(True)
plt.show()
```

Ahora nos disponemos a observar de forma similar una Serie de tiempo con tendencia Exponencial

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la serie de tiempo
n = 100  # Número de puntos de datos
growth_rate = 0.05  # Tasa de crecimiento exponencial
initial_value = 10  # Valor inicial
noise_std = 10 # Desviación estándar del ruido aleatorio
```

```python
# Generar serie de tiempo con tendencia exponencial y ruido aleatorio
t = np.arange(n)  # Tiempo
trend = initial_value * np.exp(growth_rate * t)  # Serie de tiempo exponencial
noise = np.random.normal(scale=noise_std, size=n)  # Ruido aleatorio
y = trend + noise  # Serie de tiempo exponencial con ruido
```

```python

# Graficar la serie de tiempo
plt.figure(figsize=(10, 6))
plt.plot(t, y, label='Serie de tiempo exponencial con ruido', color='red')
plt.plot(t, trend, linestyle='--', label='Tendencia exponencial', color='blue')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.title('Serie de tiempo con tendencia exponencial y ruido')
plt.legend()
plt.grid(True)
plt.show()
```

Tenemos ahora una Serie de tiempo cíclica

```python

import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la serie de tiempo
n = 100  # Número de puntos de datos
amplitude_1 = 2  # Amplitud del primer componente cíclico
frequency_1 = 0.005  # Frecuencia del primer componente cíclico
amplitude_2 = 3  # Amplitud del segundo componente cíclico
frequency_2 = 0.1  # Frecuencia del segundo componente cíclico
noise_std = 1  # Desviación estándar del ruido aleatorio


```

```python
# Generar serie de tiempo con componente cíclica y ruido aleatorio
t = np.arange(n)  # Tiempo
y = (amplitude_1 * np.cos(2 * np.pi * frequency_1 * t) +
     amplitude_2 * np.sin(2 * np.pi * frequency_2 * t) +
     np.random.normal(scale=noise_std, size=n))  # Serie de tiempo cíclica con ruido
```

Para hacer más dinámico el proceso se le agregó algo de ruido a la gráfica (algunos datos aleatorios) para que se vea mucho más realista.

```python
# Graficar la serie de tiempo
plt.figure(figsize=(10, 6))
plt.plot(t, y, label='Serie de tiempo cíclica con ruido', color='purple')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.title('Serie de tiempo con componente cíclica y ruido')
plt.legend()
plt.grid(True)
plt.show()
```

Ahora se procede a analizar una Serie de Tiempo con estacionalidad a la cuál también se le agregó ruido para que se vea un poco más real.

```python
# Generar datos de ejemplo con estacionalidad
np.random.seed(0)
n = 100  # Número de puntos de datos
t = np.arange(n)
seasonal_factor = 5  # Factor de estacionalidad
# Tendencia
trend = 0.9 * t

# Estacionalidad
seasonality = seasonal_factor * np.sin(2 * np.pi * t / 12)

# Ruido aleatorio
noise = np.random.normal(0, 1, n)

# Serie de tiempo
time_series = trend + seasonality + noise
```

```python
# Graficar
plt.figure(figsize=(10, 6))
plt.plot(t, time_series, label='Serie de tiempo', color='blue')
plt.title('Serie de tiempo con estacionalidad')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.savefig('images/STestacional.png')
plt.show()
```
