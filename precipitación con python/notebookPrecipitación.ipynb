En este cuaderno se verá un ejemplo particular de un conjunto de datos que se encuentra en csv y transformaremos sus datos que se toman diariamente a mensual, estos datos indican la precipitación diaria de 3 estaciones ubicadas en Pasco, tienen datos desde 1981 hasta 2016.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Principalmente leemos los datos correspondiente con ayuda de pandas.

```python
pcp_pasco=pd.read_csv("PCP_PASCO.csv")
pcp_pasco
```

Lo que se hace a continuación es limpiar un poco la base para quitar los indices y reemplazarlos con las fechas correspondientes.

```python
pcp_pasco["Fecha"]=pd.to_datetime(pcp_pasco.Fecha)
pcp_pasco.set_index("Fecha", inplace= True)
pcp_pasco
```

A continuación con el resample estamos creando una nueva tabla pero esta vez transformandola de datos diarios a Mensuales (rule="M"), y así obtenemos el pcp_mensual que indica la precipitación mensual en cada estación.

```python
pcp_mensual=pcp_pasco.resample(rule="M").sum()
pcp_mensual
```

Lo exportamos a un csv de nuestra preferencia

```python
pcp_mensual.to_csv("C:/Users/Hp/Desktop/precipitación con python/pcp mensual.csv")
```

Graficamos los datos diarios para la estación Yantac

```python
pcp_pasco["Yantac"].plot().set_title("Serie de precipitación diaria (mm/día) estación Yantac")
```

Y ahora observemos los datos mensuales para la estación Yantac

```python
pcp_mensual["Yantac"].plot().set_title("Serie de precipitación diaria (mm/mes) estación Yantac")
```
